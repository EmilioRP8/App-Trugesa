```python
import pandas as pd
import os
from datetime import datetime
from tabulate import tabulate

# Datos de costos y eficiencia de combustible
trailer = 2.8
rabon = 3.6
preciodecombustible = 27
combustiblemilerun = 100
porcentajeoperador = 0.13
cotizaciones = [0.10, 0.20, 0.30, 0.40]
archivo_excel = "cotizaciones.xlsx"

def calcular_cotizacion():
    import pandas as pd
    import os
    from datetime import datetime
    from tabulate import tabulate

    # Solicitar punto de arranque y destino
    lugar_partida = input("Ingrese el punto de arranque: ")
    destino = input("Ingrese el destino: ")

    # Solicitar tipo de unidad
    tipo_unidad = input("Ingrese el tipo de unidad (trailer/rabon): ").strip().lower()
    while tipo_unidad not in ['trailer', 'rabon']:
        tipo_unidad = input("Tipo de unidad invÃ¡lido. Ingrese 'trailer' o 'rabon': ").strip().lower()

    # Solicitar kilÃ³metros de ida y vuelta
    km_ida_vuelta = float(input("Ingrese los kilÃ³metros de ida y vuelta: "))

    # Calcular combustible necesario
    eficiencia = trailer if tipo_unidad == 'trailer' else rabon
    combustible_necesario = km_ida_vuelta / eficiencia

    # Preguntar si la ruta es milerun
    ruta_milerun = input("Â¿La ruta es milerun? (si/no): ").strip().lower()
    if ruta_milerun == 'si':
        combustible_necesario += combustiblemilerun  # Se suma el extra de combustible milerun

    # Calcular costo del combustible
    costo_combustible = combustible_necesario * preciodecombustible

    # Solicitar costo de las casetas
    costo_total_casetas = float(input("Ingrese el costo total de las casetas: "))

    # Calcular costos
    costo_operacion = costo_combustible + costo_total_casetas
    pago_operador = costo_operacion * porcentajeoperador
    cotizacion_costos = costo_operacion + pago_operador

    # Calcular cotizaciÃ³n final con margen del 30%
    cotizacion_final = cotizacion_costos * 1.30
    ganancia = cotizacion_final - cotizacion_costos

    # Calcular cotizaciones adicionales
    cotizaciones_valores = [cotizacion_final * (1 + cot) for cot in cotizaciones]

    # Crear un ID Ãºnico para cada cotizaciÃ³n, basado en la fecha y hora
    cotizacion_id = datetime.now().strftime("%Y%m%d%H%M%S")

    # Mostrar los datos ingresados
    # Mostrar datos ingresados y valores predefinidos
    print("\n--- DATOS INGRESADOS Y VALORES ESTABLECIDOS ---\n")
    print(f"ğŸ›£ Punto de arranque: {lugar_partida}")
    print(f"ğŸ“ Destino: {destino}")
    print(f"ğŸš› Tipo de unidad: {tipo_unidad.capitalize()}")
    print(f"ğŸ“ KilÃ³metros ida y vuelta: {km_ida_vuelta} km")
    print(f"â›½ Rendimiento ({tipo_unidad}): {eficiencia} km/l")
    print(f"ğŸ“Œ Ruta milerun: {'SÃ­' if ruta_milerun == 'si' else 'No'}")
    print(f"ğŸ”¥ Precio del combustible: ${preciodecombustible} por litro")
    print(f"â›½ Combustible necesario: {combustible_necesario:.2f} litros")
    print(f"ğŸ¢ Costo de casetas: ${costo_total_casetas}")
    print(f"ğŸ‘· Porcentaje de pago al operador: {porcentajeoperador * 100}%\n")

    # Crear DataFrame con los resultados
    columnas = ['ID CotizaciÃ³n', 'Punto de Arranque', 'Destino', 'Tipo de Unidad', 'KilÃ³metros Ida y Vuelta',
                'Ruta Milerun', 'Costo Combustible', 'Costo Casetas', 'Costo OperaciÃ³n', 'Pago Operador',
                'CotizaciÃ³n Costos', 'CotizaciÃ³n Final', 'Ganancia'] + [f'CotizaciÃ³n {int(cot*100)}%' for cot in cotizaciones]

    valores = [cotizacion_id, lugar_partida, destino, tipo_unidad, km_ida_vuelta,
               'SÃ­' if ruta_milerun == 'si' else 'No', costo_combustible, costo_total_casetas, costo_operacion, pago_operador, 
               cotizacion_costos, cotizacion_final, ganancia] + cotizaciones_valores

    df_nuevo = pd.DataFrame([valores], columns=columnas)

    # Mostrar el resultado de la cotizaciÃ³n en formato de tabla profesional
    print("\n--- RESULTADO DE LA COTIZACIÃ“N ---")
    print(tabulate(df_nuevo, headers='keys', tablefmt='fancy_grid', showindex=False))

    # Verificar si el archivo ya existe
    if os.path.exists(archivo_excel):
        # Leer el archivo existente
        df_existente = pd.read_excel(archivo_excel)
        # Agregar la nueva fila al DataFrame existente
        df_existente = pd.concat([df_existente, df_nuevo], ignore_index=True)
        # Guardar el archivo con la nueva fila
        df_existente.to_excel(archivo_excel, index=False)
    else:
        # Si el archivo no existe, crear uno nuevo con la fila
        df_nuevo.to_excel(archivo_excel, index=False)

    print(f"\nâœ… CotizaciÃ³n guardada correctamente en 'cotizaciones.xlsx'.\n")

# Asegurar que se ejecuta solo si el script es principal
if __name__ == "__main__":
    calcular_cotizacion()
```
